Recall that in PHYS 207, you were often asked to plot displacement as a function of time. For example, for a baseball thrown directly upward you would plot $y(t)$, and for a block on a horizontal spring you would plot $x(t)$. Here we will examine how the displacement and time axes "scissor together" at relativistic speeds.

Imagine that frame $K^{\prime}$ is moving in the $+x$ direction at speed $v$ relative to frame $K$. You are at rest in frame $K$. Let's figure out how the set of $(t^{\prime},x^{\prime})$ axes would look to you when the frames' origins $O$ and $O^{\prime}$ coincide.

```python
"Given that x'=0 on the t' axis we use the Lorentz transformation to find the function of x"
"0 = gamma(x-vt) which simplifies to be x = vt"
"We have x as a function of vt but we want x as a function of t' so we must use the Lorentz transformation to get there"

"Final function for x(t) that follows the t' axis is x = vt"

import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

"Defining function range"
t = np.linspace(0,1000,10)

"Need a value for the velocity of the reference frame, lets use 20 m/s"
v = 4

"Function for x(t)"
x = v * t

fig = plt.figure(figsize=(8,6))
plt.plot(t,x)
plt.xlabel('Time (seconds)')
plt.ylabel('Distance (m)')

```

    <li>Use the fact that $t^{\prime} = 0$ on the $x^{\prime}$ axis to define a line $x(t)$ that traces the $x^{\prime}$ axis in frame $K$. Write a second function that computes points on this line.</li>
</ol>

```python
"Given that t'=0 on the x' axis we use the Lorentz transformation to find the function of x"
"Lorentz transformation for time tells us t' = gamma(t- (vx)/c^2), setting that equal to 0 and solving for t we get"
"0 = gamma(t-(vx)/c^2) which simplifies to t = (vx)/c^2"
"Solving for x in the above simplified equation will yield our function for x(t) on the x' axis"

"Final function for x(t) on the x' axis ------ x = (t*c^2)/v"

import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

"Defining function range"
t = np.linspace(0,1000,10)

"Need a value for the velocity of the reference frame, lets use 20 m/s"
v = 4

"Speed of light is a np constant we can use, does not need to be defined"

"Function for x(t)"
x = (t*constants.speed_of_light**2)/v

fig = plt.figure(figsize=(6,4))
plt.plot(t,x)
plt.xlabel('Time (seconds)')
plt.ylabel('Distance (m)')
```
